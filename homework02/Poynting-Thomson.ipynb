Napětí v elastickém článku (pružině) je v případě lineárního modelu popsáno Hookeovým zákonem

$$ \sigma_{i,e}(t) = E_i \varepsilon_{i,e}(t), \ i=0,1, \tag{3.0} \label{eq:sigma_pruzina}$$

pro lineární vazký tlumič je přenášené napětí úměrné rychlosti poměrné deformace

$$ \sigma_{1,v}(t) = \eta_1 \dot{\varepsilon}_{1,v}(t). \tag{3.1} \label{eq:sigma_tlumic} $$

Při paralelním zapojení pružiny a tlumiče, které označujeme jako Kelvinův čánek, jsou oba články podrobeny stejné deformaci a celkové napětí je součtem příspěvků pružného a vazkého článku. Formálně to lze popsat vztahy

$$ \varepsilon_1(t) = \varepsilon_{1,e}(t) = \varepsilon_{1,v}(t), \tag{4.0} \label{eq:epsilon_kelvin} $$
$$ \sigma_1(t) = \sigma_{1,e}(t) + \sigma_{1,v}(t). \tag{4.1} \label{eq:sigma_kelvin} $$

Poynting-Thompson model na obr. 1 je složen z jedné pružiny a jednoho Kelvinova článku. Vzhledem k sériovému zapojení jednotek je celková poměrná deformace

$$ \varepsilon(t) = \varepsilon_0(t) + \varepsilon_1(t). \tag{5.0} \label{eq:epsilon} $$

Je zřejmé, že každá jednotka přenáší stejné celkové napětí $\sigma(t)$. Podmínky rovnováhy lze zapsat pro každou jednotku zvlášť ve tvaru

$$ \sigma(t) = \sigma_{0,e}(t), \tag{6.0} \label{eq:sigma_0} $$
$$ \sigma(t) = \sigma_{1,e}(t) + \sigma_{1,v}(t). \tag{6.1} \label{eq:sigma_1} $$

Po dosazení vztahů \eqref{eq:sigma_pruzina} a \eqref{eq:sigma_tlumic} do vztahu \eqref{eq:sigma_0} a \eqref{eq:sigma_1} dostaneme soustavu dvou navzájem nezávislých rovnic

$$ \sigma(t) = E_0 \varepsilon_{0,e}(t) \tag{7.0} \label{eq:sigma_0_dosazena}$$ 
$$ \sigma(t) = E_{1}\varepsilon_{1,e}(t) + \eta_{1} \dot{\varepsilon}_{1,v}(t). \tag{7.1} \label{eq:sigma_1_dosazena}$$

Ve vztahu \eqref{eq:sigma_0_dosazena} není nutné psát dolní index $e$, protože v pružině dochází pouze k elastické deformaci
$$ \sigma(t) = E_0 \varepsilon_{0}(t), \tag{8.0} \label{eq:sigma_0_dosazena_zjednodusena}$$ 

vztah \eqref{eq:sigma_1_dosazena} lze na základě vztahu \eqref{eq:epsilon_kelvin} zjednodušit na

$$ \sigma(t) = E_{1}\varepsilon_{1}(t) + \eta_{1} \dot{\varepsilon}_{1}(t). \tag{8.1} \label{eq:sigma_1_dosazena_zjednodusena}$$

Derivaci výrazu \eqref{eq:sigma_0_dosazena_zjednodusena} podle času lze zapsat jako

$$ \dot{\sigma}(t) = E_0 \dot{\varepsilon}_{0}(t). \tag{9.0} \label{eq:dsigma}$$

Vyjádřením $\varepsilon_1(t)$ ze vztahu \eqref{eq:epsilon} dostaneme

$$ \varepsilon_1(t) = \varepsilon(t) - \varepsilon_0(t), \tag{10.0} \label{eq:epsilon_1}$$

stejný vztah platí pro derivaci

$$ \dot{\varepsilon}_1(t) = \dot{\varepsilon}(t) - \dot{\varepsilon}_0(t). \tag{11.0} \label{eq:depsilon_1}$$

Dosazením výrazů \eqref{eq:epsilon_1} a \eqref{eq:depsilon_1} do rovnice \eqref{eq:sigma_1_dosazena_zjednodusena} dostaneme

$$ \sigma(t) = E_{1}(\varepsilon(t) - \varepsilon_0(t)) + \eta_{1} (\dot{\varepsilon}(t) - \dot{\varepsilon}_0(t)). \tag{12.0} \label{eq:sigma_1_nevimuz}$$

Rovnici lze dále upravit dosazením za $\varepsilon_0(t)$ z rovnice \eqref{eq:sigma_0_dosazena_zjednodusena} a za $\dot{\varepsilon}_0(t)$ ze vztahu \eqref{eq:dsigma}

$$ \sigma(t) = E_1 \left( \varepsilon(t) - \frac{\sigma(t)}{E_0} \right) 
+ \eta_1 \left( \dot{\varepsilon}(t) - \frac{\dot{\sigma}(t)}{E_0} \right).
\tag {13.0} \label{eq:sigma_dosazeno}
$$

Roznásobením závorek a úpravami dostaneme vztah

$$ E_1 \varepsilon(t) + \eta_1 \dot{\varepsilon}(t) = \frac{E_0 + E_1}{E_0} \sigma(t) + \frac{\eta_1}{E_0} \dot{\sigma}(t) 
\tag {14.0} \label{eq:dif_rce}
$$

## Odvození funkce poddajnosti

Model zatížíme od času $t=0\,\mathrm{s}$ konstantním napětím $\sigma(t) = \hat\sigma H(t)$, kde $H(t)$ je Heavisideova funkce daná předpisem

$$H(t) = \left\{ \begin{matrix}
0 \quad & \mathrm{pro} \ t < 0, \\
1 \quad & \mathrm{pro} \ t \geq 0.
\end{matrix}
\right.
\tag{15.0} \label{eq:Heaviside}
$$

Protože $\hat{\sigma} H(t)$ je konstantní funkce, pro její derivaci platí

$$ 
\hat{\sigma} \dot{H}(t) = 0.
\tag{16.0} \label{eq:derivace_sigma_se_striskou}
$$

Dosazením do vztahu \eqref{eq:dif_rce} dostaneme

$$ E_1 \varepsilon(t) + \eta_1 \dot{\varepsilon}(t) = \frac{E_0 + E_1}{E_0} \hat{\sigma}H(t). 
\tag {17.0} \label{eq:dif_rce_1}
$$

Rovnice \eqref{eq:dif_rce_1} je nehomogenní lineární diferenciální rovnice 1. řádu s konstantními koeficienty.

Postup řešení lineární diferenciální rovnice 1. řádu se skládá ze dvou kroků:
1. **Řešení příslušné homogenní rovnice:** Příslušnou homogenní lineární rovnici lze vyřešit separací proměnných. V jejím řešení se vyskytuje konstanta, označená $C$.
2. **Metoda variace konstanty:** Konstantu $C$ v řešení příslušné homogenní rovnice z kroku 1 považujeme za funkci proměnné $t$, tj. $$C = C(t)$$ a toto řešení (s $C(t)$ místo $C$ dosadíme do původní nehomogenní rovnice. Vznikne rovnice s jednou neznámou a to $C'=C'(t)$. Určíme $C'=C'(t)$ a integrací vypočítáme $C=C(t)$. Po dosazení za $C=C(t)$ dostaneme výsledné řešení.

Rovnici \eqref{eq:dif_rce_1} převedeme na tvar

$$ \dot{\varepsilon}(t) + \frac{E_1}{\eta_1} \varepsilon(t)  = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t). 
\tag {18.0} \label{eq:nehomogenni_dif_rce}
$$

Řešení \eqref{eq:nehomogenni_dif_rce} budeme hledat ve tvaru

$$ \varepsilon(t) = C(t) \varepsilon_h(t), \tag{19.0} \label{eq:predpokladany_tvar_reseni}$$

kde $C(t)$ je hledaná funkce proměnné $t$ a $\varepsilon_h(t)$ je řešení homogenní rovnice

$$ \dot{\varepsilon}(t) + \frac{E_1}{\eta_1} \varepsilon(t)  = 0. 
\tag {20.0} \label{eq:homogenni_dif_rce}
$$

Funkci $\varepsilon_{h}(t)$ budeme hledat ve tvaru

$$ \varepsilon_{h}(t) = C \exp{(\lambda t)}, \tag{21.0} \label{eq:tvar_reseni_epsilon}$$

kde $C$ a $\lambda$ jsou zatím neurčené konstanty.

Derivaci rovnice \eqref{eq:tvar_reseni_epsilon} lze zapsat jako

$$ \dot{\varepsilon}_{h}(t) = C \lambda \exp{(\lambda t)}.  \tag{22.0} \label{eq:tvar_reseni_depsilon}$$ 

Dosazení vztahů \eqref{eq:tvar_reseni_epsilon} a \eqref{eq:tvar_reseni_depsilon} do \eqref{eq:homogenni_dif_rce} dostaneme

$$ C \lambda \exp{(\lambda t)} + \frac{E_1}{\eta_{1}} C \exp{(\lambda t)} = 0. \tag{23.0} \label{eq:dosazeni_homo_rce} $$

Úpravou vztahu \eqref{eq:dosazeni_homo_rce} lze přejít ke tvaru

$$ C \exp{(\lambda t)} \left( \lambda + \frac{E_1}{\eta_1} \right) = 0, \tag{24.0} \label{eq:homo_rce_uprava}$$

který je splněn právě tehdy, když platí

$$ \lambda + \frac{E_1}{\eta_1} = 0. \tag{25.0} \label{eq:charakteristicka_rovnice} $$

Z této charakteristické rovnice lze vypočítat konstantu

$$ \lambda = - \frac{E_1}{\eta_1} \tag{26.0} \label{eq:lambda}. $$

Homogenní řešení má tedy tvar

$$ \varepsilon_{h}(t) = C \exp{\left(-\frac{E_1}{\eta_1}t\right)}, \tag{27.0} \label{eq:reseni_homo}$$

Položíme $C=C(t)$ a dosadíme do rovnice \eqref{eq:predpokladany_tvar_reseni}

$$ \varepsilon(t) = C(t) \exp{\left(-\frac{E_1}{\eta_1}t\right)}, \tag{28.0} \label{eq:28}$$

zderivováním výrazu dostaneme vztah

$$ \dot{\varepsilon}(t) = \dot{C}(t) \exp{\left(-\frac{E_1}{\eta_1}t\right)} - C(t)\frac{E_1}{\eta_1}\exp{\left(-\frac{E_1}{\eta_1}t\right)}.
\tag{29.0} \label{eq:derivace_epsilon}
$$

Dosazením za $\varepsilon(t)$ a $\dot{\varepsilon}(t)$ do původní nehomogenní rovnice \eqref{eq:nehomogenni_dif_rce} dostaneme vztah

$$ \dot{C}(t) \exp{\left(-\frac{E_1}{\eta_1}t\right)} - C(t)\frac{E_1}{\eta_1}\exp{\left(-\frac{E_1}{\eta_1}t\right)} + \frac{E_1}{\eta_1}C(t)\exp{\left(-\frac{E_1}{\eta_1}t\right)} = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t),
\tag{30.0} \label{eq:30}
$$

ve kterém se oba stejné členy obsahují $C(t)$ odečtou. Dostaneme tedy rovnici

$$ 
\dot{C}(t) \exp{\left(-\frac{E_1}{\eta_1}t\right)} = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t), \tag{31.0} \label{eq:31}
$$

kterou upravíme na tvar

$$ \dot{C}(t) = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t) \frac{1}{\exp{\left(-\frac{E_1}{\eta_1}t\right)}},
\tag{32.0} \label{eq:32}
$$

a dále ještě můžeme převést $\exp{\left(-\frac{E_1}{\eta_1}t\right)}$ z jmenovatele do čitatele

$$ \dot{C}(t) = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t) \exp{\left(\frac{E_1}{\eta_1}t\right)}.
\tag{33.0} \label{eq:33}
$$

Integrací obou stran rovnice \eqref{eq:33} obdržíme

$$
C(t) = \int{\frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t) \exp{\left(\frac{E_1}{\eta_1}t\right)}} \mathrm{d}t.
\tag{34.0} \label{eq:34}
$$

Po vytknutí konstant před integrál dostaneme


$$
C(t) = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t) \int{\exp{\left(\frac{E_1}{\eta_1}t\right)}} \mathrm{d}t.
\tag{35.0} \label{eq:35}
$$

Po integraci dostaneme vztah

$$
C(t) = \frac{E_0 + E_1}{E_0 \eta_1} \hat{\sigma}H(t) \frac{\eta_1}{E_1} \exp{\left(\frac{E_1}{\eta_1}t\right)} + A,
\tag{36.0} \label{eq:C}
$$

Který lze dále zjednodušit 

$$
C(t) = \frac{E_0 + E_1}{E_0 E_1} \hat{\sigma}H(t) \exp{\left(\frac{E_1}{\eta_1}t\right)} + A,
\tag{37.0} \label{eq:C_zjednodusene}
$$

kde $A$ je neznámá konstanta, kterou lze určit z počáteční podmínky

$$ \varepsilon(0) = \frac{\hat{\sigma}H(t)}{E_0}.
\tag{38.0} \label{eq:pocatecni_podminka}
$$

Dosazením $\varepsilon_h(t)$ z rovnice \eqref{eq:reseni_homo} a $C(t)$ z rovnice \eqref{eq:C_zjednodusene} do vztahu \eqref{eq:predpokladany_tvar_reseni} získáme

$$
\varepsilon(t) =
\left[
\frac{E_0 + E_1}{E_0 E_1}
\exp{\left(\frac{E_1}{\eta_1}t\right)}
\hat{\sigma} H(t) + A
\right]
\exp{\left(-\frac{E_1}{\eta_1}t\right)},
\tag{39.0} \label{eq:39}
$$

kterou lze dále upravit

$$
\varepsilon(t) =
\frac{E_0 + E_1}{E_0 E_1}
\hat{\sigma} H(t) + A \exp{\left(-\frac{E_1}{\eta_1}t\right)}.
\tag{40.0} \label{eq:40}
$$

Z počáteční podmínky \eqref{eq:pocatecni_podminka} vyplývá

$$
\frac{E_0 + E_1}{E_0 E_1}
\hat{\sigma} H(t) + A \exp{\left(0\right)} = \frac{\hat{\sigma}H(t)}{E_0}.
\tag{41.0} \label{eq:41}
$$

Úpravami lze vyřešit hodnotu konstanty $A$

$$ 
A = \frac{\hat{\sigma}H(t)}{E_0} - \frac{E_0 + E_1}{E_0 E_1} \hat{\sigma} H(t),
\tag{41.1}
$$

$$
A = \frac{\hat{\sigma} H(t) (E_1 - E_0 - E_1)}{E_0 E_1},
\tag{41.2}
$$

$$
A = \frac{- \hat{\sigma} H(t)  E_0}{E_0 E_1},
\tag{41.3}
$$

$$
A = \frac{- \hat{\sigma} H(t)}{E_1}.
\tag{41.4} \label{eq:A}
$$

Dosazením do vztahu \eqref{eq:40} dostaneme

$$
\varepsilon(t) =
\frac{E_0 + E_1}{E_0 E_1}
\hat{\sigma} H(t) - \frac{\hat{\sigma} H(t)}{E_1} \exp{\left(-\frac{E_1}{\eta_1}t\right)}.
\tag{42.0}
$$

Převedením na společného jmenovatele

$$
\varepsilon(t) =
\hat{\sigma} H(t)
\frac{E_1 + E_0 - E_0 \exp{\left(-\frac{E_1}{\eta_1}t\right)}}{E_0 E_1},
\tag{42.1}
$$

a úpravou dostaneme finální vztah pro poměrnou deformaci Poyting-Thomsonova modelu při zatěžovací zkoušce

$$
\varepsilon(t) =
\frac{\hat{\sigma} H(t)}{E_0} + \frac{\hat{\sigma} H(t)}{E_1} \left[ 1 - \exp{\left(-\frac{E_1}{\eta_1}t\right)} \right].
\tag{42.2} \label{eq:42_2}
$$

Zavedením retardačního času

$$ \tau_1 = \frac{\eta_1}{E_1}, 
\tag{43.0} \label{eq:retardacni_cas}$$

dojde ke zjednodušení vztahu \eqref{eq:42_2} na

$$
\varepsilon(t) = \left(
\frac{1}{E_0} + \frac{1}{E_1} \left[1 - \exp{\left(-\frac{t}{\tau_1}\right)} \right] \right) \hat{\sigma} H(t).
\tag{44.0} \label{eq:epsilon_zatezovaci_zkouska}
$$

Funkce poddajnosti má tedy tvar

$$
J_0(t) = \left( \frac{1}{E_0} + \frac{1}{E_1} \left[1 - \exp{\left(-\frac{t}{\tau_1}\right)} \right] \right) H(t).
\tag{45.0} \label{eq:funkce_poddajnosti}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
# koeficienty
a = 1.2
b = 1.4
c = 0.8
```

```python
E_0 = (50 + 10*a) # tuhost pruziny 0 (GPa)
E_1 = (40 + 10*b)  # tuhost pruziny 1 (GPa)
ETA_1 = (800 + 100*c) # viskozita tlumice (GPa s)

TAU_1 = ETA_1 / E_1  # retardacni cas (s)
```

```python
def calculate_J_0(time: int | float | np.ndarray) -> int | float | np.ndarray:
    return (1/E_0 + 1/E_1 * (1 - np.exp(-time/TAU_1))) * calculate_heaviside(time)

def calculate_heaviside(time: int | float | np.ndarray) -> int | float | np.ndarray:
    if isinstance(time, np.ndarray):
        heaviside = np.zeros(time.shape)
        heaviside[time >= 0] = 1
        return heaviside
    else:
        if time >= 0:
            return 1
        else:
            return 0
```

```python
t = np.linspace(-10, 150, 1000)
J_0 = calculate_J_0(t) * 10**3
H = calculate_heaviside(t)

times_of_interest = np.array([0, 10, 20, 30, np.inf])
J_0_of_interest = calculate_J_0(times_of_interest) * 10**3
```

```python
for time, compliance in zip(times_of_interest, J_0_of_interest):
    str = "J_0(t=" + f"{time:.0f}" + "\, \mathrm{s}) = " + f"{compliance:.2f}" + " \, \cdot 10^{-12} / \mathrm{Pa}"
    display(Math(str))
```

```python
def plot_point_of_interest(x, y, axes, dx, dy, ha, va):
    axes.plot(x, y, 'ko')
    label = f"[{x:.2f}, {y:.2f}]"
    axes.text(x+dx, y+dy, label, ha=ha, va=va)

fig, axs = plt.subplots(2, 1, figsize=(10,8), gridspec_kw={'height_ratios': [1, 2.5]})

axs[0].plot(t, H, 'g', label='$H(t)$')
axs[0].set_ylabel('Heaviside, $H(t)$', fontsize=16)

axs[1].plot(t, J_0, 'r', lw=2, label='$J_0(t)$')
axs[1].axhline(J_0_of_interest[-1], lw=2, ls='-.', label='$1/E_0 + 1/E_1$')
axs[1].set_ylabel('Poddajnost, $J_0(t)$ ($10^{12}$/Pa)', fontsize=16)
axs[1].set_xlabel('Čas, t (s)', fontsize=16)

for time, compliance in zip(times_of_interest[:-1], J_0_of_interest[:-1]):
    plot_point_of_interest(time, compliance, axs[1], 1, -0.5, 'left', 'top')

for i in [0, 1]:
    axs[i].grid(which="major", linestyle=(10, (20, 6)), color='black', lw=0.3)
    axs[i].minorticks_on()
    axs[i].grid(which="minor", linestyle=(0, (1, 3)), color='black', lw=0.3)
    axs[i].legend(loc="lower right", fontsize=12)
    axs[i].set_xlim(t.min(), t.max())

plt.show()
```

## Odvození relaxační funkce

Model zatížíme od času $t=0\,\mathrm{s}$ konstantním přetvořením $\varepsilon(t) = \hat\varepsilon H(t)$, kde $H(t)$ je Heavisideova funkce viz \eqref{eq:Heaviside}.

Protože $\hat\varepsilon H(t)$ je konstantní funkce, pro její derivaci platí

$$ 
\hat{\varepsilon} \dot{H}(t) = 0.
\tag{46.0} \label{eq:derivace_epsilon_se_striskou}
$$

Dosazením do vztahu \eqref{eq:dif_rce} dostaneme

$$ E_1 \hat{\varepsilon}(t) H(t) = \frac{E_0 + E_1}{E_0} {\sigma}H(t) + \frac{\eta_1}{E_0} \dot{\sigma}(t). 
\tag {47.0} \label{eq:dif_rce_47}
$$

Rovnice \eqref{eq:dif_rce_47} je nehomogenní lineární diferenciální rovnice 1. řádu s konstantními koeficienty, postup je stejný jako při odvozování funkce poddajnosti $J_0(t).$

Rovnici \eqref{eq:dif_rce_47} převedeme na tvar

$$ \dot{\sigma}(t) + \frac{E_0 + E_1}{\eta_1} \sigma(t)  = \frac{E_0  E_1}{\eta_1} \hat{\varepsilon}H(t). 
\tag {48.0} \label{eq:nehomogenni_dif_rce_sigma}
$$

Řešení \eqref{eq:nehomogenni_dif_rce_sigma} budeme hledat ve tvaru

$$ \sigma(t) = C(t) \sigma_h(t), \tag{49.0} \label{eq:predpokladany_tvar_reseni_sigma}$$

kde $C(t)$ je hledaná funkce proměnné $t$ a $\sigma_h(t)$ je řešení homogenní rovnice

$$ \dot{\sigma}_h(t) + \frac{E_0 + E_1}{\eta_1} \sigma_h(t) =  0. 
\tag {50.0} \label{eq:homogenni_dif_rce_sigma}
$$

Funkci $\sigma_{h}(t)$ budeme hledat ve tvaru

$$ \sigma_{h}(t) = C \exp{(\lambda t)}, \tag{51.0} \label{eq:tvar_reseni_sigma}$$

kde $C$ a $\lambda$ jsou zatím neurčené konstanty.

Derivaci rovnice \eqref{eq:tvar_reseni_sigma} lze zapsat jako

$$ \dot{\sigma}_{h}(t) = C \lambda \exp{(\lambda t)}.  \tag{52.0} \label{eq:tvar_reseni_dsigma}$$ 

Dosazení vztahů \eqref{eq:tvar_reseni_sigma} a \eqref{eq:tvar_reseni_dsigma} do \eqref{eq:homogenni_dif_rce_sigma} dostaneme

$$ C \lambda \exp{(\lambda t)} + \frac{E_0+E_1}{\eta_1} C \exp{(\lambda t)} = 0. \tag{53.0} \label{eq:dosazeni_homo_rce_sigma} $$

Úpravou vztahu \eqref{eq:dosazeni_homo_rce_sigma} lze přejít ke tvaru

$$ C \exp{(\lambda t)} \left( \lambda + \frac{E_0+E_1}{\eta_1} \right) = 0, \tag{54.0} \label{eq:homo_rce_uprava_sigma}$$

který je splněn právě tehdy, když platí

$$ \lambda + \frac{E_0+E_1}{\eta_1} = 0. \tag{55.0} \label{eq:charakteristicka_rovnice_sigma} $$

Z této charakteristické rovnice lze vypočítat konstantu

$$ \lambda = - \frac{E_0+E_1}{\eta_1} \tag{56.0} \label{eq:lambda_sigma}. $$

Homogenní řešení má tedy tvar

$$ \sigma_{h}(t) = C \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)}, \tag{57.0} \label{eq:reseni_homo_sigma}$$

Položíme $C=C(t)$ a dosadíme do rovnice \eqref{eq:predpokladany_tvar_reseni_sigma}

$$ \sigma(t) = C(t) \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)}, \tag{58.0} \label{eq:58}$$

zderivováním výrazu dostaneme vztah

$$ \dot{\sigma}(t) = \dot{C}(t) \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)} - C(t)\frac{E_0+E_1}{\eta_1}\exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)}.
\tag{59.0} \label{eq:derivace_sigma}
$$


Dosazením za $\sigma(t)$ a $\dot{\sigma}(t)$ do původní nehomogenní rovnice \eqref{eq:nehomogenni_dif_rce_sigma} dostaneme vztah

$$ 
\dot{C}(t) \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)} - C(t)\frac{E_0+E_1}{\eta_1}\exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)} + 
C(t)\frac{E_0+E_1}{\eta_1}\exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)} =
\frac{E_0  E_1}{\eta_1} \hat{\varepsilon}H(t)
\tag{60.0} \label{eq:60}
$$

ve kterém se oba stejné členy obsahují $C(t)$ odečtou. Dostaneme tedy rovnici

$$ 
\dot{C}(t) \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)} =
\frac{E_0  E_1}{\eta_1} \hat{\varepsilon}H(t), 
\tag{61.0} \label{eq:61}
$$

kterou upravíme na tvar

$$
\dot{C}(t) =
\frac{E_0  E_1}{\eta_1 \exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)}} \hat{\varepsilon}H(t), 
\tag{62.0} \label{eq:62}
$$

a dále ještě můžeme převést $\exp{\left(- \frac{E_0+E_1}{\eta_1} t \right)}$ z jmenovatele do čitatele

$$
\dot{C}(t) =
\frac{E_0  E_1}{\eta_1} \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)} \hat{\varepsilon}H(t).
\tag{63.0} \label{eq:63}
$$

Integrací obou stran rovnice \eqref{eq:63} obdržíme

$$
C(t) = \int{\frac{E_0  E_1}{\eta_1} \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)} \hat{\varepsilon}H(t)} \mathrm{d}t.
\tag{64.0} \label{eq:64}
$$

Po vytknutí konstant před integrál dostaneme


$$
C(t) = \frac{E_0  E_1}{\eta_1} \hat{\varepsilon}H(t) \int{ \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)}} \mathrm{d}t.
\tag{65.0} \label{eq:65}
$$

Po integraci dostaneme vztah

$$
C(t) = \frac{E_0  E_1}{\eta_1} \hat{\varepsilon}H(t) \frac{\eta_1}{E_0+E_1} \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)} + A,
\tag{66.0} \label{eq:C_sigma}
$$

který lze dále zjednodušit 

$$
C(t) = \frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t) \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)} + A,
\tag{67.0} \label{eq:C_sigma_zjednodusene}
$$

kde $A$ je neznámá konstanta, kterou lze určit z počáteční podmínky

$$ \sigma(0) = E_0 \hat{\varepsilon}H(t).
\tag{68.0} \label{eq:pocatecni_podminka_sigma}
$$

Dosazením $\sigma_h(t)$ z rovnice \eqref{eq:reseni_homo_sigma} a $C(t)$ z rovnice \eqref{eq:C_sigma_zjednodusene} do vztahu \eqref{eq:predpokladany_tvar_reseni_sigma} získáme

$$
\sigma(t) =
\left[
\frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t) \exp{\left(\frac{E_0+E_1}{\eta_1} t \right)} + A
\right]
\exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)},
\tag{69.0} \label{eq:69}
$$

kterou lze dále upravit

$$
\sigma(t) =
\frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t) + A
\exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}.
\tag{70.0} \label{eq:70}
$$

Z počáteční podmínky \eqref{eq:pocatecni_podminka_sigma} vyplývá

$$
E_0 \hat{\varepsilon}H(t) = 
\frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t) + A
\exp{\left(0\right)}.
\tag{81.0} \label{eq:81}
$$

Úpravami lze vyřešit hodnotu konstanty $A$

$$ 
A = E_0 \hat{\varepsilon}H(t) - \frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t),
\tag{71.1}
$$

$$
A = \hat{\varepsilon}H(t) \left( E_0 - \frac{E_0  E_1}{E_0+E_1} \right),
\tag{71.2}
$$

Dosazením do vztahu \eqref{eq:70} dostaneme

$$
\sigma(t) =
\frac{E_0  E_1}{E_0+E_1} \hat{\varepsilon}H(t) + 
\hat{\varepsilon}H(t) \left( E_0 - \frac{E_0  E_1}{E_0+E_1} \right)
\exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}.
\tag{72.0}
$$

Převedením na společného jmenovatele dostaneme

$$
\sigma(t) =
\hat{\varepsilon}H(t) \left(
\frac{
E_0 E_1 + E_0 (E_0 + E_1) \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)} - E_0 E_1 \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}
}{E_0 + E_1}
\right).
\tag{72.1}
$$

Dále lze vytknout před závorku $\frac{E_0}{E_0+E_1}$

$$
\sigma(t) =
\hat{\varepsilon}H(t) \frac{E_0}{E_0 + E_1} \left(
E_1 + (E_0 + E_1) \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)} - E_1 \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}
\right).
\tag{72.2}
$$

Sečtením výrazů v závorce dostaneme finální vztah pro napětí Poyting-Thomsonova modelu při deformační zkoušce

$$
\sigma(t) =
\hat{\varepsilon}H(t) \frac{E_0}{E_0 + E_1} \left(
E_1 + E_0 \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}
\right).
\tag{72.3} \label{eq:72_3}
$$

Relaxační funkce má tedy tvar

$$
R_0(t) = 
H(t) \frac{E_0}{E_0 + E_1} \left(
E_1 + E_0 \exp{\left(-\frac{E_0+E_1}{\eta_1}t\right)}
\right).
\tag{73.0} \label{eq:relaxacni_funkce}
$$

```python
def calculate_R_0(time: int | float | np.ndarray) -> int | float | np.ndarray:
    return calculate_heaviside(time) * (E_0 / (E_0 + E_1) * (E_1 + E_0 * np.exp(-time * (E_0 + E_1) / ETA_1)))
```

```python
t = np.linspace(-10, 150, 1000)
R_0 = calculate_R_0(t)
H = calculate_heaviside(t)

times_of_interest = np.array([0, 10, 20, 30, np.inf])
R_0_of_interest = calculate_R_0(times_of_interest)
```

```python
for time, relaxation in zip(times_of_interest, R_0_of_interest):
    str = "R_0(t=" + f"{time:.0f}" + "\, \mathrm{s}) = " + f"{relaxation:.4f}" + " \, \mathrm{GPa}"
    display(Math(str))
```

## Porovnání $R_0(t) = \frac{1}{J_0(t)}$

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))

axs[0].plot(t[R_0!=0], 1/R_0[R_0!=0] * 10**3, 'r', lw=2, label='$1/R_0(t)$')
axs[0].plot(t, J_0, 'g', lw=2, label='$J_0(t)$')
axs[0].axhline(J_0_of_interest[-1], lw=2, ls='-.', label='$1/E_0 + 1/E_1$')
axs[0].set_ylabel('($10^{12}$/Pa)', fontsize=16)

axs[1].plot(t, R_0, 'r', lw=2, label='$R_0(t)$')
axs[1].plot(t[J_0!=0], 1/J_0[J_0!=0] * 10**3, 'g', lw=2, label='$1/J_0(t)$')
axs[1].axhline(R_0_of_interest[-1], lw=2, ls='-.', label='$E_0 E_1 / (E_0 + E_1)$')
axs[1].set_ylabel('(GPa)', fontsize=16)
axs[1].set_xlabel('Čas, t (s)', fontsize=16)

for i in [0, 1]:
    axs[i].grid(which="major", linestyle=(10, (20, 6)), color='black', lw=0.3)
    axs[i].minorticks_on()
    axs[i].grid(which="minor", linestyle=(0, (1, 3)), color='black', lw=0.3)
    axs[i].legend(loc="lower right", fontsize=12)
    axs[i].set_xlim(t.min(), t.max())
```

Z grafu je patrné, že vztah 

$$ R_0(t) = \frac{1}{J_0(t)}
\tag{74.0} \label{eq:R_is_J}
$$

platí pouze pro $t = 0 \, \mathrm{s}$ a pro $t \rightarrow +\infty$.

Ověříme dosazením pro $t = 0 \, \mathrm{s}$ do vztahu \eqref{eq:funkce_poddajnosti} 

$$
J_0(0) = \left( \frac{1}{E_0} + \frac{1}{E_1} \left[1 - \exp{\left(-\frac{0}{\tau_1}\right)} \right] \right) H(0) = 
\left( \frac{1}{E_0} + \frac{1}{E_1} \left[1 - \exp{0} \right] \right) \cdot 1 = \frac{1}{E_0},
\tag{75.0} \label{eq:J0_0}
$$

a do \eqref{eq:relaxacni_funkce}

$$
R_0(0) = 
H(0) \frac{E_0}{E_0 + E_1} \left(
E_1 + E_0 \exp{\left(-\frac{E_0+E_1}{\eta_1} \cdot 0 \right)} 
\right) = 
1 \cdot \frac{E_0}{E_0 + E_1} \left( E_1 + E_0 \exp{(0)}  \right) =
\frac{E_0(E_0 + E_1)}{E_0 + E_1} = E_0.
\tag{76.0} \label{eq:R0_0}
$$

To samé provedeme pro $t \rightarrow +\infty$.

$$
J_0(t \rightarrow +\infty) = \left( \frac{1}{E_0} + \frac{1}{E_1} \left[1 - \exp{\left(-\frac{t \rightarrow +\infty}{\tau_1}\right)} \right] \right) H(t \rightarrow +\infty) = 
\left( \frac{1}{E_0} + \frac{1}{E_1} \left[1 - 0 \right] \right) \cdot 1 = \frac{1}{E_0} + \frac{1}{E_1} = \frac{E_0+E_1}{E_0 E_1},
\tag{77.0} \label{eq:J0_oo}
$$


$$
R_0(t \rightarrow +\infty) = 
H(t \rightarrow +\infty) \frac{E_0}{E_0 + E_1} \left(
E_1 + E_0 \exp{\left(-\frac{E_0+E_1}{\eta_1} \cdot t \rightarrow +\infty \right)} 
\right) = 
1 \cdot \frac{E_0}{E_0 + E_1} \left( E_1\right) =
\frac{E_0 E_1}{E_0 + E_1}.
\tag{78.0} \label{eq:R0_oo}
$$

Z výsledků je zřejmé, že vztah \eqref{eq:R_is_J} platí pro $t = 0 \, \mathrm{s}$ a pro $t \rightarrow +\infty$.

## Popis vývoje deformace při lineární rostoucím napětí

Předpokládejme lineární růst napětí z nulové hodnoty v čase $t_0 = 0 \, \mathrm{s}$ až do hodnoty $\sigma(t_1) = \hat{\sigma}$.

Funkci napětí lze tedy vyjádřit vztahem

$$
\sigma(t) = \frac{\hat{\sigma}}{t_1}t,
\tag{79.0}
$$

a její derivaci

$$
\dot{\sigma}(t) = \frac{\hat{\sigma}}{t_1}.
\tag{80.0}
$$

V tomto případě se poměrná deformace vypočítá podle vztahu

$$
\varepsilon(t) = \int_{t_0}^t J(t, t') \dot{\sigma}(t') \mathrm{d}t
\tag{81.0} \label{eq:pomerna_deformace}
$$

Dosazením za $J(t,t')$ a $\dot{\sigma}(t)$ dostaneme rovnici

$$
\varepsilon(t) = \int_0^t 
\left(
\frac{1}{E_0} +
\frac{1}{E_1}
\left[
1 - \exp \left( -\frac{t-t'}{\tau_1} \right)
\right]
\right)
\frac{\hat{\sigma}}{t_1}
\mathrm{d}t'.
\tag{81.0}
$$

Rozdělením na dílčí integrály a vytknutím konstant před integrál dostaneme

$$
\varepsilon(t) =
\frac{\hat{\sigma}}{t_1}
\left[
\frac{1}{E_0} \int_0^t \mathrm{d}t' +
\frac{1}{E_1} \int_0^t \mathrm{d}t' -
\frac{1}{E_0} \int_0^t \exp \left( -\frac{t-t'}{\tau_1} \right) \mathrm{d}t'
\right].
\tag{81.1}
$$

Integrací získáme

$$
\varepsilon(t) = 
\frac{\hat{\sigma}}{t_1}
\left[
\frac{1}{E_0} [t']_0^t +
\frac{1}{E_1} [t']_0^t -
\frac{1}{E_1} \tau_1 \exp \left( -\frac{t}{\tau_1} \right) \left[ \exp{ \left( \frac{t'}{\tau_1} \right) - 1} \right]_0^t
\right].
\tag{81.2}
$$

Dosazením dostaneme výraz pro poměrnou deformaci 

$$
\varepsilon(t) =
\frac{\hat{\sigma}}{t_1} 
\left[
\frac{t}{E_0} +
\frac{t}{E_1} -
\frac{\tau_1}{E_1} +
\frac{\tau_1}{E_1} \exp{\left( \frac{-t}{\tau_1} \right)}
\right].
\tag{81.3}
$$

Při extrémně rychlém zatížení ($t \rightarrow 0^+)$ se deformuje pouze pružina, vztah pro deformaci tedy lze napsat jako

$$
\varepsilon = \frac{\hat{\sigma}}{E_0}.
\tag{82.0}
$$

Při extrémně rychlém zatížení ($t \rightarrow \infty)$ se rychlost deformace tlumiče blíží nule ($\dot{\varepsilon}_v \rightarrow 0$) a model se chová jako dvě pružiny zapojené sériově, lze tedy psát

$$
\varepsilon = \hat{\sigma}\left( \frac{1}{E_0} + \frac{1}{E_1} \right).
\tag{83.0}
$$

```python
SIGMA_HAT = 8  # (GPa)
def stress(t: int| float | np.ndarray, t_1: int | float) -> int | float | np.ndarray:
    return (SIGMA_HAT / t_1) * t

def strain(t: int | float | np.ndarray, t_1: int | float) -> int | float | np.ndarray:
    return (SIGMA_HAT / t_1) * (t/E_0 + t/E_1 - TAU_1/E_1 + TAU_1/E_1 * np.exp(-t/TAU_1))
```

```python
t_1a = 10
t_1b = 20
t_1c = 40


t_a = np.linspace(0, t_1a, 100)
t_b = np.linspace(0, t_1b, 100)
t_c = np.linspace(0, t_1c, 100)

sigma_a = stress(t_a, t_1a)
sigma_b = stress(t_b, t_1b)
sigma_c = stress(t_c, t_1c)

strain_a = strain(t_a, t_1a)
strain_b = strain(t_b, t_1b)
strain_c = strain(t_c, t_1c)

strain_0 = SIGMA_HAT / E_0  # extremne rychle zatizeni
strain_oo = SIGMA_HAT * ( 1/E_0 + 1/E_1)  # extremne pomale zatizeni
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))

axs[0].plot(t_a, sigma_a, label='$t_1 = 10 \, \mathrm{s}$')
axs[0].plot(t_b, sigma_b, label='$t_1 = 20 \, \mathrm{s}$')
axs[0].plot(t_c, sigma_c, label='$t_1 = 40 \, \mathrm{s}$')
axs[0].set_ylabel('Napětí, $\sigma$ (MPa)', fontsize=12)
axs[0].set_title('Vývoj napětí v čase', fontsize=16)

axs[1].plot(t_a, strain_a, label='$t_1 = 10 \, \mathrm{s}$')
axs[1].plot(t_b, strain_b, label='$t_1 = 20 \, \mathrm{s}$')
axs[1].plot(t_c, strain_c, label='$t_1 = 40 \, \mathrm{s}$')
axs[1].set_ylabel('Poměrné přetvoření, $\\varepsilon$ ($\cdot 10^{-3}$)', fontsize=12)
axs[1].set_xlabel('Čas, t (s)', fontsize=12)
axs[1].set_title('Vývoj poměrné deformace v čase', fontsize=16)

for i in [0, 1]:
    axs[i].grid(which="major", linestyle=(10, (20, 6)), color='black', lw=0.3)
    axs[i].minorticks_on()
    axs[i].grid(which="minor", linestyle=(0, (1, 3)), color='black', lw=0.3)
    axs[i].legend(loc="lower right", fontsize=12)
    axs[i].set_xlim(0, t_1c)
```
